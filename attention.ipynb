```python
import torch.nn.functional as F

def attention_block(query: torch.Tensor, key: torch.Tensor, values: torch.Tensor) -> torch.Tensor:
    dk = query.shape[-1]
    qk = (query @ key.transpose(-2,-1))/(dk**0.5)
    qk_sm = F.softmax(qk, dim=-1)
    return qk_sm @ values
```
