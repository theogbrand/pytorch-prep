## 1D + 2D -> 2D
"Single Axis broacasting"

```python
t_rand = torch.rand((3,)) # Shape 1,3
print(t_rand)
```

```python
t_rand.shape
```

```python
t_ones = torch.ones((4,1))
print(t_ones)
```

```python
len(t_ones.shape)
```

```python
torch.add(t_rand, t_ones)
```

```python
torch.add(t_ones, t_rand)
```

```python
t = torch.tensor([3])
print(t.item())
```
