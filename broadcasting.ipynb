Multi-Axis Broadcasting

# torch sum(), torch.max(), torch.eq()

```python
# torch.max(), torch.sum()
A = torch.tensor([[0,1,2], [0,2,3]]) # 2D tensor
B = torch.tensor([[4,0,5], [6,0,7]]) # 2D tensor
torch.sum(A==0, dim=0)
```

```python
torch.sum(torch.eq(A, 0), 0)
```

```python
zeros_per_col = torch.sum(A==0, dim=0)
```

```python
torch.max(torch.sum(A==0, dim=0))
```

```python
max_num_0s = torch.max(zeros_per_col)
```

```python
A[:,zeros_per_col != max_num_0s]
```

```python
zeros_per_col_b = torch.sum(torch.eq(B, 0), dim=0)
torch.argmax(zeros_per_col_b)
```

```python
col_max_num_zeros_b = torch.argmax(zeros_per_col_b)
torch.cat([B[:, :col_max_num_zeros_b], B[:, col_max_num_zeros_b+1:]], dim=1)
```

These all fail because the shapes are not broadcastable. (after attempting to broadcast on both tensors)<br>
